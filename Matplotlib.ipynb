## Making Figures with Numpy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Hostogram 
* Subplots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,10, 100)
```

```python
x[: 10]
```

```python
#Plot data and create line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make scatter plot
fig, ax= plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax= plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
#Make a plot from dictionary
nut_butter_prices = {"Almond butter" : 10,
                    "Peanut butter" : 8,
                    "Cashew butter" : 12}
fid, ax= plt.subplots()
ax.bar(nut_butter_prices.keys(),height=nut_butter_prices.values())
# in python we can't declare first one feature while plotting something
ax.set(title="Dan's Nut Butter Store",
      ylabel= "Prices ($) ");
```

```python
# for bar horizontal
fig, ax= plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
# Make some data for histogram and plot it
x= np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for sub-plots

```python
# Sub-plot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols= 2,
                                            figsize= (10,5))
#Plot to each different figures
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(10000));#randn produce random normalized graph
```

```python
#Subplot option 2
fig, ax= plt.subplots(nrows= 2,
                       ncols=2,
                      figsize=(10, 5))
#plot to each different index,
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one to use (pyplot vs matplotlib oo method)

* when plotting something quickly use pyplot method
* when plotting something advanced use matplotlib OO method

```python
heart_disease
```

```python
over_50= heart_disease[heart_disease["age"] > 50]
```

```python
over_50
```

```python
len(over_50)
```

```python
over_50.head()
```

```python
over_50.plot.scatter(x= "age",
                    y= "chol",
                    c= "target",
                    figsize=(10,6));
```

```python
#OO method
fig, ax= plt.subplots(figsize= (10,6))
over_50.plot.scatter(x= "age",
                     y= "chol",
                     c= "target",
                     ax=ax);

ax.set_xlim([45,100])
```
